# 1. Imports and environment setup

```python
import pandas as pd
import numpy as np
from joblib import load
import warnings
```

# 2. Load Data and Model

```python
testDataFrame= pd.read_csv('../data/preprocessed_test.csv', delimiter=',')
```

```python
testDataFrame.shape
```

```python
testDataFrame.describe()
```

```python
# Load the selected features from the file
selected_features = load('../data/selected_features.joblib')
```

```python
# Load the model from the file
best_model = load('../data/best_random_forest_model.joblib')
```

# 3. Prediction

## 3.1. Make Predictions

```python
# Suppress the warning
warnings.filterwarnings("ignore", message="X has feature names, but RandomForestClassifier was fitted without feature names")

# Drop the 'Status' column
testDataFrame.drop(columns=['Status'], inplace=True)

# Select only the columns corresponding to the selected features
testDataFrame = testDataFrame[selected_features]

# Make predictions
predictions = best_model.predict(testDataFrame)

# Convert predictions to 0s and 1s based on the condition
predictions_binary = [0 if pred == 0 else 1 for pred in predictions]

# Convert predictions to a DataFrame
predictions_df = pd.DataFrame({'Status': predictions_binary})
```

## 3.2. Preview of Predictions

```python
predictions_df.info()
```

```python
predictions_df
```

# 4. Save the Predictions to the file

```python
# Save predictions to CSV
predictions_df.to_csv("../data/prediction/COSC2753_A1_s3939114.csv", index=False)
```
